```python
# module import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data import
usa = pd.read_excel('./USA_DATA.xlsx', encoding = 'utf-8')
usa.head()
```

```python
# The relationship between human capital and political institution
fig, ax1 = plt.subplots(figsize=(15, 7))

ax1.plot(usa['Year'], usa['Average Years of Education'], 'ro-')
ax2 = ax1.twinx()                                            # x축 공유
ax2.plot(usa['Year'], usa['Political Participation'], 'yD-')


plt.title('USA relationship between human capital and political institution', pad=15)
ax1.set_xlabel('Year', labelpad=15)
ax1.set_ylabel('Average Years of Education(●)', labelpad=15)
ax2.set_ylabel('Political Participation(◆)', labelpad=15)

plt.grid()
#plt.show()

plt.savefig('4.png')
```

```python
# data import
usa2 = pd.read_excel('./USA_DATA_2.xlsx', encoding = 'utf-8')
usa2.head()
```

```python
# The relationship between human capital and the standard of living
fig, ax1 = plt.subplots(figsize=(15, 7))

ax1.plot(usa2['Year'], usa2['Numeracy (Total)'], 'go-')
ax2 = ax1.twinx()                                            # x축 공유
ax2.plot(usa2['Year'], usa2['Height'], 'cD-')


plt.title('USA relationship between human capital and the standard of living', pad=15)
ax1.set_xlabel('Year', labelpad=15)
ax1.set_ylabel('Numeracy(●)', labelpad=15)
ax2.set_ylabel('Height(◆)', labelpad=15)

plt.grid()
#plt.show()

plt.savefig('2_4.png')
```
