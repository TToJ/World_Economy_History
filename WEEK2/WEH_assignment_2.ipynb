```python
# module import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data import
germany = pd.read_excel('./Germany_Data.xlsx', encoding = 'utf-8')
germany.head()
```

```python
# GDP per Capita graph
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot()

ax.plot(germany['Year'], germany['GDP per Capita'], 'g.-')
plt.grid()

plt.title('GDP per Capita of Germany', pad=15)
plt.xlabel('Year', labelpad=15)
plt.ylabel('GDP per Capita', labelpad=15)
#plt.show()

plt.savefig('GDP.png')
```

```python
# GDP per Capita + Heigth graph
fig, ax1 = plt.subplots(figsize=(18, 7))

ax1.plot(germany['Year'], germany['GDP per Capita'], 'g.-')
ax2 = ax1.twinx()                                            # x 
ax2.plot(germany['Year'], germany['Height'], 'm.-')


plt.title('Germany GDP per Capita & Height', pad=15)
ax1.set_xlabel('Year', labelpad=15)
ax1.set_ylabel('GDP per Capita', labelpad=15)
ax2.set_ylabel('Height', labelpad=15)

plt.grid()
#plt.show()

plt.savefig('GDP_Height.png')
```
