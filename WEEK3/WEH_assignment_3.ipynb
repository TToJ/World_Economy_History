```python
# module import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data import
germany_pop = pd.read_excel('./Germany_Data_Population.xlsx', encoding = 'utf-8')
germany_pop.head()
```

```python
# Total Pop + Infant Mortality
fig, ax1 = plt.subplots(figsize=(15, 7))

ax1.plot(germany_pop['Year'], germany_pop['Total Population'], 'go-')
ax2 = ax1.twinx()                                            # x축 공유
ax2.plot(germany_pop['Year'], germany_pop['Infant Mortality'], 'mD-')


plt.title('Germany Total Population & Infant Mortality', pad=15)
ax1.set_xlabel('Year', labelpad=15)
ax1.set_ylabel('Total Population(●)', labelpad=15)
ax2.set_ylabel('Infant Mortality(◆)', labelpad=15)

plt.grid()
#plt.show()

plt.savefig('population.png')
```
